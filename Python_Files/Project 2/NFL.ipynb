```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from decimal import Decimal
from collections import Counter


file_to_load = "Resources/2018-2019NFLSeason.csv"
file_to_load1 = "Resources/2018-2019NHLSeason.csv"
file_to_load2 = "Resources/NBA2018-2019Season.csv"
file_to_load3 = "Resources/2018MLBSeason.csv"
file_to_load4 = "Resources/2018MLSSeason.csv"

# Read Purchasing File and store into Pandas data frame
NFLSeason_Table = pd.read_csv(file_to_load)
NHLSeason_Table = pd.read_csv(file_to_load1)
NBASeason_Table = pd.read_csv(file_to_load2)
MLBSeason_Table = pd.read_csv(file_to_load3)
MLSSeason_Table = pd.read_csv(file_to_load4)
```

```python
df=pd.DataFrame(NFLSeason_Table)
df1=pd.DataFrame(NHLSeason_Table)
df2=pd.DataFrame(NBASeason_Table)
df3=pd.DataFrame(MLBSeason_Table)
df4=pd.DataFrame(MLSSeason_Table)
df.head(2)
```

```python
df = df[df.home_score.notnull()]
df.tail()
```

```python
dfh=df['home_team']
```

```python
Team_appearences= pd.value_counts(df["home_team"])+pd.value_counts(df["away_team"])
Team_appearences.head(8)
```

```python
teams = list(set(df.home_team.values))
wins = []
py_wins = []
total_points_for_list = []
total_points_against_list = []
```

```python
for team in teams:
    data_home = df[df.home_team == team]
    data_home['win'] = np.where(data_home.home_result > 0, 1, 0)
    home_count = data_home.home_team.count()
    win_home = data_home.win.sum()
    points_for_home = data_home.home_score.sum()
    
    points_against_home = data_home.away_score.sum()
    data_away = df[df.away_team == team]
    data_away['win'] = np.where(data_away.home_result < 0, 1, 0)
    away_count = data_away.away_team.count()
    win_away = data_away.win.sum()
    points_for_away = data_away.away_score.sum()
    points_against_away = data_away.home_score.sum()
    
    total_points_for = int(points_for_home + points_for_away)
    total_points_for_list.append(total_points_for)
    total_points_against = int(points_against_home + points_against_away)
    total_points_against_list.append(total_points_against)
    
#     total_games = home_count + away_count
    total_wins = (win_home + win_away)
    wins.append(total_wins)
   
```

```python
NFL_Table= pd.DataFrame(list(zip(teams, total_points_for_list, total_points_against_list, wins)), columns =['Team', 'Points For', 'Points Against', 'Wins'])  
NFL_Table.head(10)
```

```python
Teams_Clean = Team_appearences.reset_index().rename(columns={'index':'Team',0:'Games'})
NFL_Clean=NFL_Table.merge(Teams_Clean, on="Team" )
NFL_Clean['Wins %']=NFL_Clean["Wins"]/NFL_Clean["Games"]
pd.options.display.float_format = "{:,.2f}".format
NFL_Clean
```

```python
NFL_Clean.to_csv('NFL_UPDATED.csv',index=False)
```
